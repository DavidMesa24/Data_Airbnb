```python
frec_data=data['review_scores_rating'].value_counts().reset_index()
frec_data.head(10)
```

```python
frec_data['Frecuencia_relativa']=frec_data['review_scores_rating'].apply(lambda x: x/frec_data['review_scores_rating'].sum())
frec_data.head(5)
```

```python
frec_data['frec_rel_acum']=frec_data['Frecuencia_relativa'].apply(lambda x: x*100)
frec_data.head(5)
```

```python
frec_data['Frec_acum'] = frec_data['review_scores_rating'].cumsum()
frec_data.head(5)
```

```python
frec_data['Frec_acum_porc']=frec_data['frec_rel_acum'].cumsum()
frec_data.head(5)
```
